# Popularity Based Recommendation Engine

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
rating_num_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
rating_num_df.rename(columns={'Book-Rating':'Rating-Num'}, inplace=True)
```

```python
rating_avg_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
rating_avg_df.rename(columns={'Book-Rating':'Rating-Avg'}, inplace=True)
```

```python
rating_df = rating_num_df.merge(rating_avg_df, on='Book-Title')
popular_df = rating_df[rating_df['Rating-Num'] > 250].sort_values('Rating-Avg', ascending=False).head(50)
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title').drop(columns = ['ISBN','Image-URL-S', 'Image-URL-M'])
popular_df.head(50)
```

## Collaborative filter based recommender system

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 150
learned_users = x[x].index
learned_users
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(learned_users)]
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating'] > 50
filtered_books = y[y].index
filtered_df = filtered_ratings[filtered_ratings['Book-Title'].isin(filtered_books)]
filtered_df
```

```python
pt = filtered_df.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_table = cosine_similarity(pt)
```

```python
similarity_table[0]
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    distances = list(enumerate(similarity_table[index]))
    sorted_distances = sorted(distances, key=lambda x:x[1], reverse=True)
    recoms = sorted_distances[1:11]
    rec_books = []
    
    for i in recoms:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Publisher'].values))
        rec_books.append(item)
    
   
    return rec_books
```

```python
recommend('1984')
```
